```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.read_csv('predictions/prdiction_1.csv') 
df2 = pd.read_csv('predictions/prdiction_2.csv') 
df3 = pd.read_csv('predictions/prdiction_3.csv') 
df4 = pd.read_csv('predictions/prdiction_4.csv') 
```

```python
df2 = pd.merge(df1[["id"]],df2,on="id",how='left')
df3 = pd.merge(df1[["id"]],df3,on="id",how='left')
df4 = pd.merge(df1[["id"]],df4,on="id",how='left')
df1
```

```python
df1["target"] = (df1["target"]*0.25)+(df2["target"]*0.25)+(df3["target"]*0.25)+(df4["target"]*0.25)
```

```python
df1.to_csv('prdiction.csv',index=False)
df1
```
